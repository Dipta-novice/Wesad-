```python
import pickle
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import csv
```

```python
s2p="/content/drive/MyDrive/S4.pkl"
with open(s2p, 'rb') as file:
    s2_data = pickle.load(file, encoding='latin1')
```

Extraction of Wrist Data

```python
w_ax=s2_data['signal']['wrist']['ACC']
w_ay=s2_data['signal']['wrist']['ACC']
w_az=s2_data['signal']['wrist']['ACC']
w_bvp=s2_data['signal']['wrist']['BVP']
w_eda=s2_data['signal']['wrist']['EDA']
w_temp=s2_data['signal']['wrist']['TEMP']
w_label=s2_data['label']
```

No of Data in each property

```python
%matplotlib inline
print("w_ax",len(w_ax))
print("w_ay",len(w_ay))
print("w_az",len(w_az))
print("w_bvp",len(w_bvp))
print("w_eda",len(w_eda))
print("w_temp",len(w_temp))
print(w_label, len(w_label))


fig = plt.figure()
ax = fig.add_axes([1,1,2,2])
xaxis = ["w_ax","w_ay","w_az","w_bvp","w_eda","w_temp","w_label"]
y_ax = [len(w_ax),len(w_ay),len(w_az),len(w_bvp),len(w_eda),len(w_temp),len(w_label)]
ax.bar(xaxis,y_ax,color = 'c', width = 0.45)

plt.show()
```

Here I have made dataframes for each property with the labels till its count

```python
numpy_data1=np.array([w_eda,w_label[0:25692]])
numpy_data2=np.array([w_ax,w_label[0:205536]])
numpy_data3=np.array([w_ay,w_label[0:205536]])
numpy_data4=np.array([w_az,w_label[0:205536]])
numpy_data5=np.array([w_temp,w_label[0:25692]])
numpy_data1=numpy_data1.T
numpy_data2=numpy_data2.T
numpy_data3=numpy_data3.T
numpy_data4=numpy_data4.T
numpy_data5=numpy_data5.T


dfweda = pd.DataFrame(data=numpy_data1, columns=["w_eda","Label"])
dfwax = pd.DataFrame(data=numpy_data2, columns=["w_ax","Label"])
dfway = pd.DataFrame(data=numpy_data3, columns=["w_ay","Label"])
dfwaz = pd.DataFrame(data=numpy_data4, columns=["w_az","Label"])
dfwtemp = pd.DataFrame(data=numpy_data5, columns=["w_temp","Label"])
numpy_data=np.array([w_bvp,w_label[0:411072]]) 
numpy_data=numpy_data.T
df_wbvp = pd.DataFrame(data=numpy_data, columns=['WBVP','Label'])

```

```python
dfweda.to_csv('WedaS4.csv')
dfwax.to_csv('WaxS4.csv')
dfway.to_csv('WayS4.csv')
dfwaz.to_csv('WazS4.csv')
dfwtemp.to_csv('WtempS4.csv')
df_wbvp.to_csv('WwbvpS4.csv')
```
